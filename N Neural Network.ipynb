```python
def tanh(z):
    return np.tanh(z)
```

```python
def derivative_tanh(z):
    return (1-(np.tanh(z))**2)
```

```python
def relu(z):
    return (np.maximum(0,z))
```

```python
def derivative_relu(z):
    return (z>0)
```

```python
def sigmoid(z):
    answer = 1/(1+np.exp(-z))
    return answer
```

```python
def derivative_sigmoid(z):
    answer = sigmoid(z)*(1-sigmoid(z))
    return answer
```

```python
def softmax(z):
    num=np.exp(z)
    den=np.sum(num,axis=1)
    den=np.expand_dims(den,axis=1)
    soft=num/den
    return soft
```

```python
def cross_entropy_loss(y,a):
    answer = (1/m)*np.sum(y*np.log(a))
    return -answer
```
